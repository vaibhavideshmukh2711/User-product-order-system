```python
class User:
    def __init__(self):
        self.uid = None
        self.mobile = None
        
    def signup(self,mobile,passkey):
        try:
            cursor = conn.cursor(buffered=True)
            cursor.execute("")
            cursor.execute("insert into user1(mobile,passkey) values (%s,%s)",(mobile,passkey))
            conn.commit()
            print("Signup successfull")
            cursor.close()
        except mysql.connector.IntegrityError:
            print("Mobile already exists")
            
    def signin(self,mobile,passkey):
        cursor = conn.cursor(buffered=True)
        cursor.execute("select uid from user1 where mobile = %s and passkey = %s",(mobile,passkey))
        user = cursor.fetchone()
        cursor.close()
        if user:
            self.uid = user[0]
            self.mobile = mobile
            print("Login successfull")
            return True
        else:
            print("Invalid details")
            return False
        
    def view_products(self):
        cursor = conn.cursor(buffered=True)
        cursor.execute("select * from product")
        products = cursor.fetchall()
        cursor.close()
        print("Avialable products ") 
        for p in products:
            print(f"PID: {p[0]}, Name: {p[1]}")
            
    def place_order(self,pid):
        if not self.uid:
            print("please login first")
            return 
        try:
            cursor = conn.cursor(buffered=True)
            cursor.execute("SELECT pname FROM product WHERE pid=%s", (pid,))
            product = cursor.fetchone()
            
            if not product:
                print("product not found")
                cursor.close()
                return
            
            product_name = product[0]
            cursor.execute("insert into orders(uid,pid,product_name) values (%s,%s,%s)",(self.uid,pid,product_name))
            conn.commit()
            cursor.close()
            print("orders are placed ")
        except mysql.connector.Error as e:
            print("Error inserting order:",e)             
              # get the name from the tuple
        # Insert into orders table with product name
                          
    def view_orders(self):
        if not self.uid:
            print("Print login first")
            return 
        cursor = conn.cursor(buffered=True)
        cursor.execute("select o.oid,p.pname from orders o join product p on o.pid = p.pid where o.uid = %s",(self.uid,))
        orders = cursor.fetchall()
        cursor.close()
        print("Your order:")
        for o in orders:
            print(f"Order ID: {o[0]},Product:{o[1]}")
```
